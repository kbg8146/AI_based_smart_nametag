```python
#  Install & Imports (Colab only)
!pip install --upgrade gspread gspread_dataframe oauth2client

import matplotlib.pyplot as plt
from google.colab import auth
import pandas as pd
import gspread
from gspread_dataframe import get_as_dataframe
from google.auth import default
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages

#  Google authentication
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

#  Open spreadsheet and worksheet
spreadsheet_url = "https://docs.google.com/spreadsheets/d/1kkt336f1G-XqfDuwCUOnqpKlxTcnwLQy-XS4SQv6lM0/edit"
sh = gc.open_by_url(spreadsheet_url)
ws = sh.worksheet("Clustered Result with Distance")

#  Load dataframe
df = get_as_dataframe(ws, evaluate_formulas=True).dropna(subset=[" "])

#  Rename Korean industry columns to English
korean_to_english = {
    "HW": "HW Design",
    " /": "Semiconductor",
    " SW": "Embedded SW",
    "/RF": "Comm./RF",
    "/": "Control/Robot",
    "AI": "AI",
    "": "Digital Circuit",
    "": "Analog Circuit",
    "": "Automotive"
}
df.rename(columns=korean_to_english, inplace=True)

#  English industry labels
industries = list(korean_to_english.values())

#  Assign cluster color
cluster_colors = [
    "blue", "green", "orange", "purple", "red", "brown", "pink", "gray", "cyan", "olive"
]

#  Radar plot function
def plot_radar_chart(scores, labels, name, group_name, cluster_id, ax=None):
    N = len(labels)
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    scores = scores + scores[:1]
    angles = angles + angles[:1]

    if ax is None:
        fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

    color = cluster_colors[int(cluster_id) % len(cluster_colors)]
    ax.plot(angles, scores, color=color, linewidth=2)
    ax.fill(angles, scores, color=color, alpha=0.4)

    ax.set_title(f"{name} (Cluster {cluster_id})", size=12, y=1.08)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, fontsize=9)
    ax.set_yticklabels([])
    ax.grid(True)

#  Save all radar charts by cluster to PDF
pdf_filename = "radar_charts_by_cluster_grouped_fixed.pdf"
with PdfPages(pdf_filename) as pdf:
    grouped = df.groupby("Cluster")
    for cluster_id, group_df in grouped:
        num_participants = len(group_df)
        rows = int(np.ceil(num_participants / 3))
        fig, axs = plt.subplots(nrows=rows, ncols=3, figsize=(15, 5 * rows), subplot_kw=dict(polar=True))
        axs = axs.flatten()

        for i, (_, row) in enumerate(group_df.iterrows()):
            name = row[""] if "" in row else f"Participant {i+1}"
            group_name = row.get("Cluster_Name", "")
            try:
                scores = [float(row[col]) if not pd.isna(row[col]) else 0 for col in industries]
                normalized_scores = [s / 100 for s in scores]
                plot_radar_chart(normalized_scores, industries, name, group_name, int(cluster_id), ax=axs[i])
            except Exception as e:
                print(f" Error plotting for {name}: {e}")

        # Remove unused subplots
        for j in range(i + 1, len(axs)):
            fig.delaxes(axs[j])

        fig.suptitle(f"Cluster {int(cluster_id)}  Participant Radar Charts", fontsize=16)
        pdf.savefig(fig)
        plt.close(fig)

print(f" PDF saved as: {pdf_filename}")



```
